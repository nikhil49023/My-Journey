```python
import pandas as pd
```

Series
A  pandas 1-D array, that can hold any type of data

```python
data = [100, 2 , 300]
series = pd.Series(data)
print(series)
```

INDEX is used to pass the index values for the data in series (S.No)
length of data must be equal to labels

```python
series = pd.Series(data, index =["a", "b", "c"])
print(series)
```

Loc : Location By Lable
Iloc : Location By Index
it will return the value of the locaton

```python
print(series.loc["a"])
print(series.iloc[2])
```

```python
print(series[series <= 100])
#RETURNING CONDITIONAL OUTPUTS { <, >, =, !}
```

Python Dictionaries are key-value pairs, so keys act as labels and valuse act as series

```python
dictionary = {"nikhil": 500, "sai": 900, "ram": 800}
series = pd.Series(dictionary)
print(series)
```

DATA FRAMES

A TABULAR DATA STRUCTURE WITH ROWS AND COLUMNS. (2-DIMENSIONAL)

```python
data = {"Name": ["raju", "rohit", "ravi", "nikhil"], "Age": [15, 18, 36, 19]}
data_frame = pd.DataFrame(data, index=[1,2,3,4])
print(data_frame)
```

```python
data_frame["Job"] = ["Gamer", "Traveler", "Front-End Dev", "Hacker"]
print(data_frame)
#TO ADD A COLUMN TO DATA FRAME
```

```python
new_row = pd.DataFrame([{"Name" : "Manoj", "Age": 17, "Job": "Fighter"}])
data_frame = pd.concat([data_frame, new_row], )
print(data_frame)
#TO ADD A ROW TO DATA FRAME
```
